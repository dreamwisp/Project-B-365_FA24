```python
path = 'B365-HW6.csv'
```

```python
import pandas as pd
```

```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(path)
```

```python
df.info()
```

```python
colors = [
    "#8A2BE2",  
    "#FF6347",  
    "#6A5ACD",  
    "#FFD700",  
    "#7FFFD4",  
    "#EE82EE",  
    "#FF4500",  
    "#40E0D0",  
    "#9370DB", 
    "#3CB371",  
]
len(colors)
```

```python
sse = []
k_list = [i for i in range(1,21)]

print(k_list) 
for i in range(1,10):
    sse = []
    for k in range(1,21):
        kmeans = KMeans(n_clusters=k, init ='random', n_init='auto').fit(df)
        sse.append(kmeans.inertia_)
    plt.plot(k_list,sse)
    plt.xticks(k_list)
    #plt.plot(k_list, sse, c=colors[i % len(colors)]) 
    #plt.figure(figsize=(7,5)) 
    #plt.plot(k_list,sse, c= colors[i-1])
    
plt.xlabel('Number of Clusters')
plt.ylabel("SSE")
plt.title("SSE of Clusters using K-means")
plt.show()
```
